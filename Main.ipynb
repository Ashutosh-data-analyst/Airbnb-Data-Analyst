```python
# How Price Depend on Neighbourhood
sns.barplot(data=df, x="neighbourhood_group", y="price", hue="room_type")
plt.show()
```

```python
# Number of review and Price relations
plt.figure(figsize=(8,5))
plt.title("Location and Review Dependency")
plt.xlabel("No. of Reviews")
plt.xlabel("Price")
sns.scatterplot(data=df,x='number_of_reviews', y="price", hue="neighbourhood_group")
plt.show()
```

```python
#Pair chart
sns.pairplot(data=df, vars=['price','minimum_nights','availability_365', 'number_of_reviews'], hue="room_type")
plt.show()
```

```python
#Pair chart 
# GEO chart based on Latitude and longitude
plt.figure(figsize=(8,8))
sns.scatterplot(data=df,x="longitude",y="latitude", hue="room_type")
plt.title("Geographical Distribution of Airbnb Listing")
plt.show()
```

```python
# Heat Map
corr = df[["latitude", "longitude", "price", "minimum_nights", "number_of_reviews", "reviews_per_month", "availability_365", "beds"]].corr()
#cor = corr.head(200)

plt.figure(figsize=(8,6))
sns.heatmap(data=corr , annot=True, fmt="0.2f") 
plt.show()
```
